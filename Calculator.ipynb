```python
import math

history = []

def save_history(entry):
    history.append(entry)
    with open("history.txt", "a") as file:
        file.write(entry + "\n")


while True:

    print("\n===== Advanced Python Calculator =====")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power")
    print("6. Square Root")
    print("7. sin(x)")
    print("8. cos(x)")
    print("9. log(x)")
    print("10. factorial")
    print("11. Evaluate Expression")
    print("12. Show History")
    print("13. Exit")

    choice = input("Choose option (enter number): ")

    try:

        # EXIT
        if choice == "13":
            print("Goodbye!")
            break

        # SHOW HISTORY
        elif choice == "12":
            print("\n--- History ---")
            for item in history:
                print(item)
            input("\nPress Enter to continue...")

        # EXPRESSION CALCULATOR
        elif choice == "11":
            expr = input("Enter expression (example: 5+3*2): ")
            result = eval(expr)
            entry = f"{expr} = {result}"
            print(entry)
            save_history(entry)
            input("\nPress Enter to continue...")

        # SINGLE NUMBER FUNCTIONS
        elif choice in ["6","7","8","9","10"]:

            num = float(input("Enter number: "))

            if choice == "6":
                result = math.sqrt(num)
                entry = f"√{num} = {result}"

            elif choice == "7":
                result = math.sin(num)
                entry = f"sin({num}) = {result}"

            elif choice == "8":
                result = math.cos(num)
                entry = f"cos({num}) = {result}"

            elif choice == "9":
                result = math.log(num)
                entry = f"log({num}) = {result}"

            elif choice == "10":
                result = math.factorial(int(num))
                entry = f"{int(num)}! = {result}"

            print(entry)
            save_history(entry)
            input("\nPress Enter to continue...")

        # TWO NUMBER OPERATIONS
        elif choice in ["1","2","3","4","5"]:

            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))

            if choice == "1":
                result = num1 + num2
                entry = f"{num1} + {num2} = {result}"

            elif choice == "2":
                result = num1 - num2
                entry = f"{num1} - {num2} = {result}"

            elif choice == "3":
                result = num1 * num2
                entry = f"{num1} * {num2} = {result}"

            elif choice == "4":
                if num2 == 0:
                    print("Cannot divide by zero")
                    input("\nPress Enter to continue...")
                    continue
                result = num1 / num2
                entry = f"{num1} / {num2} = {result}"

            elif choice == "5":
                result = num1 ** num2
                entry = f"{num1}^{num2} = {result}"

            print(entry)
            save_history(entry)
            input("\nPress Enter to continue...")

        else:
            print("Invalid option")
            input("\nPress Enter to continue...")

    except Exception as e:
        print("Error:", e)
        input("\nPress Enter to continue...")
```
